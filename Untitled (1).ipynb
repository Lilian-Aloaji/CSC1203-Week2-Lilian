```python
# Example 3:
num = int(input("Enter a number: "))
if num>0:
    print("Positive")
else:
    print("Not Positive")
if num>=10 and num<=50:
    print("The number is between 10 and 50")
else:
    print("The number is NOT between 10 and 50")
```

```python
# Example 4:
color = input("Enter a color (red, yellow, green): ")
if color != "":
   match color.lower():
    case "red":
        print("Stop")
    case "yellow":
        print("Get Ready")
    case "green":
        print("Go")
    case _:
        print("Invalid color")
else:
    print("Input cannot be empty")
```

```python
# Examble 5:
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = input("Choose an option (1-4): ")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

match choice:
    case "1":
        print("Result:", num1 + num2)
    case "2":
        print("Result:", num1 - num2)
    case "3":
        print("Result:", num1 * num2)
    case "4":
        if num2 != 0:
            print("Result:", num1 / num2)
        else:
            print("Cannot divide by zero")
    case _:
        print("Invalid choice")
```

```python
# Examble 6:
score = int(input("Enter student score: "))

if score >= 50:
    print("Passed")
else:
    print("Failed")

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 50:
    grade = "D"
else:
    grade = "F"

match grade:
    case "A":
        print("Excellent")
    case "B":
        print("Very Good")
    case "C":
        print("Good")
    case "D":
        print("Pass")
    case "F":
        print("Fail")
```
